```python
#Part 1 - Rock, Paper, Scissors (continued)
#Defining ___ function

def computer_brain(games, rock, paper, scissors):
    global computer_pick
    
    paper_prob = paper / games
    rock_prob = rock / games
    scissor_prob = scissors / games
    
    if paper_prob > rock_prob and paper_prob > scissor_prob:
        computer_pick = "scissors"
    elif rock_prob > paper_prob and rock_prob > scissor_prob:
        computer_pick = "paper"
    elif scissor_prob > paper_prob and scissor_prob > rock_prob:
        computer_pick = "rock"
    elif paper_prob > rock_prob and paper_prob == scissor_prob:
        number = random.randint(1, 2)
        if number == 1:
            computer_pick = "rock"
        elif number == 2:
            computer_pick = "scissors"
    elif scissor_prob > paper_prob and scissor_prob == rock_prob:
        number = random.randint(1, 2)
        if number == 1:
            computer_pick = "rock"
        elif number == 2:
            computer_pick = "paper"
    elif paper_prob > scissor_prob and paper_prob == rock_prob:
        number = random.randint(1, 2)
        if number == 1:
            computer_pick = "paper"
        elif number == 2:
            computer_pick = "scissors"
    
    return computer_pick
```

```python
#Part 3 - Near-Duplicate Detection (continued)
#Defining moving_window and cosine functions

def moving_window(tweet):
    #Split the words in the tweet
    #.split() stores the words in a list--tweet_list
    tweet_list = tweet.split(' ')
    
    #Create an empty string
    empty_string = ''
    
    #Creates an empty dictionary
    dict_list = {}
    
    length = len(tweet_list)
    
    #Iterate through the first 3 indexes of the tweet_list, add to empty_string
    #Store empty_string as the key in a dictionary with value = 1
    #Continue to loop until the end of the string
    while length >= 3:
        empty_string = tweet_list[0] + ' ' + tweet_list[1] + ' ' + tweet_list[2]
        dict_list[empty_string] = 1
        tweet_list.pop(0)
        length -= 1
        empty_string = ''

    return dict_list

def cosine(tweet1, tweet2):
    n1 = len(tweet1)
    n2 = len(tweet2)
    denom = math.sqrt(n1 * n2)
    
    key_list1 = list(tweet1.keys())
    key_list2 = list(tweet2.keys())
    
    match = [phrase for phrase in key_list1 if phrase in key_list2]
    match_num = len(match)
    
    cos = match_num / denom
    
    return cos
```
